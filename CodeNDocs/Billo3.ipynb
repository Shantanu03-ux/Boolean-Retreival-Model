```python
# defining soundex algorithm
def soundex(name):
    s = []
    si = 1
    c = 0
    #           ABCDEFGHIJKLMNOPQRSTUVWXYZ
    mappings = "01230120022455012623010202";

    s.append(name[0].upper())
    Len = len(name)
    for i in range(1,Len):
        c = ord(name[i].upper()) - 65
        if (c >= 0 and c <= 25):
            if mappings[c] != '0':
                if mappings[c] != s[si-1]:
                    s.insert(si,mappings[c])
                    si+=1

                if si > 3:
                    break

    if si <= 3:
        while si <= 3:
            s.insert(si,'0')
            si+=1
        

    return s
```
